```python
import tensorflow as tf
 
import pandas as pd
 
import matplotlib.pyplot as plt
%matplotlib inline
 
# Import Keras libraries
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
df = pd.read_csv('diabetes.csv')
```

```python
df
```

```python
x_input = df.drop('Outcome', 1).values
y_label = df[['Outcome']].values
```

```python
x_input
```

```python
x_input.shape
```

```python
y_label
```

```python
y_label = tf.keras.utils.to_categorical(y_label)
```

```python
y_label
```

```python
from sklearn.utils import shuffle
```

```python
x_input, y_label = shuffle(x_input, y_label, random_state=42)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    x_input, y_label, test_size=0.10, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
model = Sequential()
model.add(Dense(units=20, input_dim=8, activation = 'relu'))
#model.add(Dropout(0.3))
model.add(Dense(units=10, activation = 'relu'))
model.add(Dropout(0.5))
model.add(Dense(units=2, activation = 'softmax'))
```

```python
model.compile(optimizer='sgd', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train, y_train, epochs=1000, validation_data=(X_test, y_test))
```

```python
x_input.shape
```

```python
model.evaluate(X_test, y_test)
```
