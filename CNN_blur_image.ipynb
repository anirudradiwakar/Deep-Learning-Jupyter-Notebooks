```python
import tensorflow as tf
```

```python
import matplotlib.pyplot as plt
```

```python
from scipy import misc
```

```python
img = misc.ascent()
```

```python
plt.figure(figsize=(5,5))
```

```python
plt.imshow(img, cmap='gray')
```

```python
img.shape
```

```python
img_tensor = img.reshape((1, 512, 512, 1))
```

```python
model = tf.keras.Sequential()
```

```python
from tensorflow.keras.layers import Conv2D
```

```python
model.add(Conv2D(filters=1, kernel_size=(3, 3), kernel_initializer='ones', input_shape=(512,512,1)))
```

```python
model.compile(optimizer='adam', loss='mse')
```

```python
model.summary()
```

```python
img_predict_tensor = model.predict(img_tensor)
```

```python
img_predict_tensor.shape
```

```python
plt.imshow(img_predict_tensor.reshape((img_predict_tensor.shape[1], img_predict_tensor.shape[2])), cmap='gray')
```
