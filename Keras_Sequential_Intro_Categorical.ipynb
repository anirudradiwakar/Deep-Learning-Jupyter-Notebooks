```python
import tensorflow as tf
 
import pandas as pd
 
import matplotlib.pyplot as plt
%matplotlib inline
 
# Import Keras libraries
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
df = pd.read_csv('data.csv')
```

```python
plt.scatter(df[df['label'] == 0]['x1'], df[df['label'] == 0]['x2'], marker='*')
plt.scatter(df[df['label'] == 1]['x1'], df[df['label'] == 1]['x2'], marker='<')
```

```python
x_input = df[['x1','x2']].values
y_label = df[['label']].values
```

```python
x_input
```

```python
y_label
```

```python
y_label = tf.keras.utils.to_categorical(y_label)
```

```python
y_label
```

```python
model = Sequential()
model.add(Dense(units=2, input_dim=2, activation = 'softmax'))
```

```python
model.compile(optimizer='sgd', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x_input, y_label, epochs=1000)
```

```python
x_input.shape
```

```python
model.evaluate(x_input, y_label)
```

```python
import numpy as np
```

```python
pred = model.predict(np.reshape([12.0, 16.0], (1, 2)))
```

```python
pred
```

```python
if(np.argmax(pred)==0):
    print("Class A")
    
else:
    print("Class B")
```
